```python
grouped_ratings = df.groupby('movieId')['rating'].agg(['count', 'mean']).reset_index()
grouped_ratings.columns = ['movieId', 'ratings_count', 'ratings_mean']
merged_data = pd.merge(df, grouped_ratings, on='movieId', how='inner')
filtered_movies = merged_data[merged_data['ratings_count'] > 50]
print(filtered_movies)
```

```python
average_ratings = filtered_movies.groupby('title')['rating'].mean().reset_index()

# Find the movie with the highest average rating
most_popular_movie = average_ratings.loc[average_ratings['rating'].idxmax()]
most_popular_movie
```
